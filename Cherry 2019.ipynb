```python
import pandas as pd
```

```python
df= pd.read_csv('Cherry_2019.csv')
df
```

```python
feature = ['state','district','market','variety','arrival_date','min_price','max_price']
x = df[feature]# Features
y = df.modal_price# Target variable
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

df.state_encoded=le.fit_transform(df.state)
print("state" ,df.state_encoded)
df.district_encoded=le.fit_transform(df.district)
print("district" ,df.district_encoded)
df.market_encoded=le.fit_transform(df.market)
print("market" ,df.market_encoded)
df.variety_encoded=le.fit_transform(df.variety)
print("variety" ,df.variety_encoded)
df.arrival_date_encoded=le.fit_transform(df.arrival_date)
print("arrival_date" ,df.arrival_date_encoded)
```

```python
x=list(zip(df.state_encoded,df.district_encoded,df.market_encoded,df.variety_encoded,df.arrival_date_encoded,df.min_price,df.max_price))
print(x)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
# Calculate correlations
corr = df.corr()

# Heatmap
sns.heatmap(corr)
```

```python
from sklearn.model_selection import train_test_split# Import train_test_split function

# Split dataset into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state=100) 
# 70% training and 30% test
```

```python
# import the class
from sklearn.linear_model import LinearRegression
# fit the model with data
regr = LinearRegression()
regr.fit(x_train,y_train)
# accuracy calculation
accuracy= regr.score(x_test,y_test)
print(accuracy*100,'%')
```
